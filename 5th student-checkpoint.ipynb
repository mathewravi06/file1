```python
import matplotlib.pyplot as plt
import pandas as pd
data={"name":["john","vihan","krishna","arthi"],
      "maths":[70,75,96,87],
      "science":[95,50,65,90],
      "social":[50,95,56,60],
      "english":[50,85,89,67],
      "hindi":[59,73,64,91]}
df=pd.DataFrame(data)
print(df)
pivot=df.melt(id_vars="name",var_name="subject",value_name="score")
pivot_transposed=df.transpose()
print(pivot_transposed)
print("adding new row")
new_row={"name":"new student","maths":70,"science":60,"social":45,"english":60,"hindi":85}
df=df._append(new_row,ignore_index=True)
df["subject"]=[100,90,80,70,60]
df_transposed = df.transpose()
print(df_transposed)
print("sortig maths values")
sorted_df=df.sort_values(by="maths",ascending=False)
print(sorted_df)
#plot code
plt.figure(figsize=(10,6))
plt.bar(df["name"],df["maths"],label="maths")
plt.bar(df["name"],df["science"],label="science")
plt.bar(df["name"],df["social"],label="social")
plt.bar(df["name"],df["english"],label="english")
plt.bar(df["name"],df["hindi"],label="hindi")
plt.legend()
plt.show()
#line plot
plt.figure(figsize=(10,6))
plt.plot(df["name"],df["maths"],marker="o",label="maths")
plt.plot(df["name"],df["science"],marker="o",label="science")
plt.plot(df["name"],df["social"],marker='o',label="social")
plt.plot(df["name"],df["english"],marker="o",label="english")
plt.plot(df["name"],df["hindi"],marker='o',label="hindi")
plt.legend()
plt.show()
```
