```python
import pandas as pd

# Creating the initial dataframe
data = {
    'Column1': [10, 11, 12, 13, 14],
    'Column2': [23, 34, 45, 32, 65],
    'Column3': [55, 60, 65, 70, 75]
}
df = pd.DataFrame(data)

# (a) Display the dataframe
print("Initial Dataframe:")
print(df)

# Adding the new list as a column
df['Column4'] = [1, 2, 3, 4, 5]

# (b) Display the updated dataframe
print("\nDataframe after adding the new list:")
print(df)
```

```python
import pandas as pd

# Creating the series
names = pd.Series(['Alice', 'Bob', 'Charlie', 'David', 'Eva'])
grades = pd.Series(['A', 'B', 'C', 'B', 'A'])
marks = pd.Series([85, 78, 92, 70, 88])

# Creating the dataframe from the series
student_df = pd.DataFrame({
    'Name': names,
    'Grade': grades,
    'Marks': marks
})

# (a) Display the first three records
print("First three records:")
print(student_df.head(3))

# (b) Display the last two records
print("\nLast two records:")
print(student_df.tail(2))
```
